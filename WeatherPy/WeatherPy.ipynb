```python
#Import Depencecies

import numpy as np
import pandas as pd
from citipy import citipy
import matplotlib.pyplot as plt
import requests as req
import json
import scipy.stats as st
```

```python
#Import API key and construct request url

from api_keys import api_key
temp_units = "imperial"
query_url = (f"http://api.openweathermap.org/data/2.5/weather?appid={api_key}&units={temp_units}&q=")
```

```python
#Set lists for first loop
lat_lst = []
lon_lst = []
coord_lst = []
cit_lst = []

#Set lists for second loop
cit_cln = []
cit_country = []
cit_lat = []
cit_lng = []
cit_max_temp = []
cit_humidity = []
cit_cloudiness = []
cit_wind_speed = []

#Set blank DF
cit_weather_df = pd.DataFrame()
```

```python
#create random coordinates list using numpy random function for use against citypy. 180 degrees latitude 360 degrees longitude 

lat_lst = np.random.uniform(low=-90, high=90, size=1500)
lon_lst = np.random.uniform(low=-180, high=180, size=1500)
coord_lst = zip(lat_lst,lon_lst)

#Search nearest cities to coordinates in citypy and append to list

for coord_lst in coord_lst:
    city = citipy.nearest_city(coord_lst[0], coord_lst[1]).city_name
    if city not in cit_lst:
        cit_lst.append(city)

print(cit_lst)
```

```python
len(cit_lst)
```

```python
#set city count for printing in loop

cit_count = 0

#Loop request using city list and append results into lists for merging into a dictionary then a dataframe

for city in cit_lst:
    try:
        response = req.get(query_url + city.replace(" ","&")).json()
        cit_cln.append(response['name'])
        cit_country.append(response['sys']['country'])
        cit_lat.append(response['coord']['lat'])
        cit_lng.append(response['coord']['lon'])
        cit_max_temp.append(response['main']['temp'])
        cit_humidity.append(response['main']['humidity'])
        cit_cloudiness.append(response['clouds']['all'])
        cit_wind_speed.append(response['wind']['speed'])
        print(f'Scanning: {cit_count} of {len(cit_lst)} | {city} added as number {len(cit_cln)}')
        cit_count = cit_count + 1
    
    except KeyError:
        print(f'No Entry {cit_count} of {len(cit_lst)}. {city} not added')
        cit_count = cit_count + 1
        
print("Scanning Complete")
```

```python
#Import lists to dictionary and convert to Pandas Dataframe

cit_weather = {
    "City":cit_cln,
    "Country":cit_country,
    "Latitude":cit_lat,
    "Longitude":cit_lng,
    "Temperature-Max-Farenheit":cit_max_temp,
    "Humidity-Percentage":cit_humidity,
    "Cloudiness-Percentage": cit_cloudiness,
    "Wind-Speed":cit_wind_speed
}

city_weather_df = pd.DataFrame.from_dict(cit_weather)

#Check first 30 entries for data integrity

city_weather_df.head(30)
```

```python
#Split Datasets into Southern and Northern Hemispheres

sh_weather_df = city_weather_df[city_weather_df.Latitude < 0]
nh_weather_df = city_weather_df[city_weather_df.Latitude > 0]
```

# Temperature Vs Latitude - Northern Vs Southern Hemisphere

```python
#Plot Temperature vs Latitude for Southern Hemisphere 

(slope, intercept, rvalue, pvalue, stderr) = st.linregress(sh_weather_df['Latitude'], sh_weather_df['Temperature-Max-Farenheit'])
regress_values = sh_weather_df['Latitude'] * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

plt.figure(figsize=(12,10))
plt.scatter(sh_weather_df['Latitude'], sh_weather_df['Temperature-Max-Farenheit'])
plt.plot(sh_weather_df['Latitude'],regress_values,"r-")
plt.annotate(line_eq,(12,10),fontsize=15,color="red")
plt.xlabel('Latitude (Degrees)')
plt.ylabel('Temperature (°F)')
plt.title(f'Latitude vs. Max Temperature for a Random Selection of Cities in The Southern Hemisphere')
plt.grid(color='k', linestyle='--', linewidth=1)
print(f"The r-squared is: {rvalue**2}")
print(f"The Linear Regression model is: {line_eq}")
plt.show()
#Save plot as *.*png

plt.savefig('output/SHtemperature.png')
```

```python
#Plot Temperature vs Latitude for Northern Hemisphere

(slope, intercept, rvalue, pvalue, stderr) = st.linregress(nh_weather_df['Latitude'], nh_weather_df['Temperature-Max-Farenheit'])
regress_values = nh_weather_df['Latitude'] * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

plt.figure(figsize=(12,10))
plt.scatter(nh_weather_df['Latitude'], nh_weather_df['Temperature-Max-Farenheit'])
plt.plot(nh_weather_df['Latitude'],regress_values,"r-")
plt.annotate(line_eq,(12,10),fontsize=15,color="red")
plt.xlabel('Latitude (Degrees)')
plt.ylabel('Temperature (°F)')
plt.title(f'Latitude vs. Temperature for a Random Selection of Cities in The Northern Hemisphere')
plt.grid(color='k', linestyle='--', linewidth=1)
print(f"The r-squared is: {rvalue**2}")
print(f"The Linear Regression model is: {line_eq}")
plt.show()

#Save plot as *.*png

plt.savefig('output/NHtemperature.png')
```

 <b> Summary </b> <br>
   The above show a strong correlation between latitude and temperature. In the southern hemisphere it is a positive correlation as temperature increases as you move towards the equator. In the northern hemisphere it is the inverse as temperature decreases as you move away from the equator.
